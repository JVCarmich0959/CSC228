# Using the DistilBERT model

The DistilBERT model is a small, fast, cheap and light Transformer model trained by distilling Bert base. It has 40% less parameters than bert-base-uncased, runs 60% faster while preserving over 95% of Bertâ€™s performances as measured on the GLUE language understanding benchmark.

It is a good choice for a first-time user because of its speed and performance.

```python
nlp = pipeline("question-answering", model="distilbert-base-cased-distilled-squad", tokenizer="distilbert-base-cased")

context = "Our mall has a variety of stores, including clothing stores, electronics stores, and specialty stores."
question = "What stores are in the mall?"
result = nlp(question=question, context=context)
print(result)
```

```python
# Using the BERT model
nlp = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad", tokenizer="bert-large-uncased-whole-word-masking-finetuned-squad")

context = "Our mall has a variety of stores, including clothing stores, electronics stores, and specialty stores."
question = "What stores are in the mall?"
result = nlp(question=question, context=context)
print(result)
```

```python
def get_answer(question, context):
    # Use the DistilBERT QA pipeline to get the answer
    result = qa_pipeline(question=question, context=context)
    
    # Format the output
    answer = result['answer']
    score = result['score']
    formatted_answer = f"Answer: {answer}\nScore: {score:.2f}"
    
    # Return the formatted answer
    return formatted_answer


# function to get answer from the BERT model
def get_answer_bert(question, context):
    # tokenize the question and context
    inputs = bert_tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt")
    
    # get the start and end indices of the answer from the model output
    start_scores, end_scores = bert_model(**inputs).values()
    answer_start = torch.argmax(start_scores)
    answer_end = torch.argmax(end_scores) + 1
    
    # get the answer text from the token IDs
    answer = bert_tokenizer.convert_tokens_to_string(bert_tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end]))
    
    # return the answer
    return answer

# function to get answer from the RoBERTa model
def get_answer_roberta(question, context):
    # tokenize the question and context
    inputs = roberta_tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt")
    
    # get the start and end indices of the answer from the model output
    start_scores, end_scores = roberta_model(**inputs).values()
    answer_start = torch.argmax(start_scores)
    answer_end = torch.argmax(end_scores) + 1
    
    # get the answer text from the token IDs
    answer = roberta_tokenizer.convert_tokens_to_string(roberta_tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end]))
    
    # return the answer
    return answer
```

```python
for index, row in df.iterrows():
    # get the question and context from the data frame
    question = row['question']
    context = row['context']
    
    # get the answer from the DistilBERT model
    answer_distilbert = get_answer(question, context)
    
    # get the answer from the BERT model
    answer_bert = get_answer(question, context)
    
    # get the answer from the RoBERTa model
    answer_roberta = get_answer(question, context)
    
    # update the data frame with the answers
    df.at[index, 'answer_distilbert'] = answer_distilbert
    df.at[index, 'answer_bert'] = answer_bert
    df.at[index, 'answer_roberta'] = answer_roberta
    
df
```
