```python
# Read custom Sentiment Dataset here
import pandas as pd 

data = pd.read_csv('https://raw.githubusercontent.com/JVCarmich0959/CSC228/main/WUG_Mall_improvement.csv',usecols=[0,1,2])
data.head()
```

**Sentiment Analysis of your Dataset**

Using the utterances from your Dataset and the sentiment-analysis pipeline from Hugging Face, create a predict column in your Dataframe with the Hugging Face prediction of sentiment.

Calculate the number of correct vs incorrect predictions (your golden/expert labels are the "right answer")

```python
# Add a column named predict and score to the dataframe
data['predict'] =  data['Sentence'].apply(lambda x: classifier(x)[0]['label'])
data['score'] =  data['Sentence'].apply(lambda x: classifier(x)[0]['score'])

# Change the predict column to a numeric value
data['predict'] = data['predict'].replace({'NEGATIVE': 0, 'POSITIVE': 1})

# if the predict value and the Label value are the same, then the prediction is correct
data['correct'] = data['predict'] == data['Label']

```

```python
#print data
data.head()
```

```python
# calculate the number of correct vs incorrect predictions
correct = data['correct'].value_counts()[1]
incorrect = data['correct'].value_counts()[0]

print('Number of correct predictions: ', correct)
print('Number of incorrect predictions: ', incorrect)

# calculate the accuracy and print as a percentage
accuracy = correct / (correct + incorrect)
print('Accuracy: ', accuracy * 100, '%')
```

**Bottom N Accurate Predictions**

Write function(s) to find the N least accurate POSITIVE sentiments (per your golden labels)

Repeat for the N least accurate NEGATIVE sentiments 

You may write one or more functions.

```python
import pandas as pd

def find_least_accurate_positive_predictions(data, n):
    # create a dataframe with only the positive predictions
    positive = data[data['predict'] == 1]

    # sort the positive dataframe by the score column
    positive = positive.sort_values(by=['score'], ascending=True)

    # return the top N least accurate positive predictions as a list of tuples (sentence, score)
    return list(zip(positive['Sentence'].head(n), positive['score'].head(n)))

def find_least_accurate_negative_predictions(data, n):
    # create a dataframe with only the negative predictions
    negative = data[data['predict'] == 0]

    # sort the negative dataframe by the score column
    negative = negative.sort_values(by=['score'], ascending=True)

    # return the top N least accurate negative predictions as a list of tuples (sentence, score)
    return list(zip(negative['Sentence'].head(n), negative['score'].head(n)))



```

```python
# find the 5 least accurate positive predictions
print(find_least_accurate_positive_predictions(data, 5))

# find the 5 least accurate negative predictions
print(find_least_accurate_negative_predictions(data, 5))
```

**Top N Accurate Predictions**

Write function(s) to find the N most accurate POSITIVE sentiments (per your golden labels)

Repeat for the N most accurate NEGATIVE sentiments 

You may write one or more functions.

```python
# Write your function(s) here.
def find_most_accurate_predictions(data, n, sentiment):
    if sentiment == 'POSITIVE':
        # create a dataframe with only the positive predictions
        predictions = data[data['predict'] == 1]
    elif sentiment == 'NEGATIVE':
        # create a dataframe with only the negative predictions
        predictions = data[data['predict'] == 0]
    else:
        raise ValueError("Sentiment must be 'POSITIVE' or 'NEGATIVE'")

    # sort the predictions dataframe by the score column
    predictions = predictions.sort_values(by=['score'], ascending=False)

    # return the top N most accurate predictions as a list of tuples (sentence, score)
    return list(zip(predictions['Sentence'].head(n), predictions['score'].head(n)))
```

```python
# Run your function here for N=5 for POSITIVE SENTIMENTS

# find the 5 most accurate positive predictions
print(find_most_accurate_predictions(data, 5, 'POSITIVE'))
```

```python
# Run your function here for N=5 for NEGATIVE SENTIMENTS

# find the 5 most accurate negative predictions
print(find_most_accurate_predictions(data, 5, 'NEGATIVE'))
```
